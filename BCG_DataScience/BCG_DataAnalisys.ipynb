## Overall comparison of Companies

From the data it is clearly visible that AAPL is having the biggest figures in these three years. MSFT have only bigger Assets than Apple but it gives the way in all others categories. TSLA as rising company is not yet on level of Microsoft or Apple  which proves enormous gap between two market giants and rising Tesla.

```python
import matplotlib.pyplot as plt
columns_to_plot = ["Total revenue($)", "Net income($)", "Total Assets($)", \
"Total Liabilities($)", "Cash Flow($)"]
#groups data by company and sum metrics
df_grouped = df.groupby('Company')[columns_to_plot].sum()

df_grouped_in_billions = df_grouped / 1e9

ax = df_grouped_in_billions.plot(kind = "bar", stacked = True, figsize = (14, 7))

plt.title("General graph sorted by Comapny")
plt.xlabel("Company")
plt.ylabel("Financial figures($) in billions from 3 years")

plt.legend(title = "Metrics", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.grid(axis = "y", linestyle = "--", alpha = 0.7)
plt.tight_layout()

plt.show()

```

## Breaking down the data

### Total Revenue

Microsoft has shown consistent revenue growth, increasing from $198 billion in 2022 to $245 billion in 2024, reflecting a 24% rise. Apples revenue remained relatively stable, reaching $391 billion in 2024 after a slight dip in 2023. Tesla, while having significantly lower revenue compared to the other two, demonstrated rapid growth, increasing from $53 billion in 2021 to $96 billion in 2023, indicating its expansion efforts in a highly competitive market.

```python
plt.figure(figsize = (10, 6))

for company in df["Company"].unique():
    company_data = df[df["Company"] == company]
    plt.plot(company_data["Year"], company_data["Net income($)"] / 1e9, \
label = company, marker = "o")

    for i in range(len(company_data)):
        plt.text(company_data["Year"].iloc[i], \
company_data["Total revenue($)"].iloc[i] / 1e9, \
f'{company_data["Total revenue($)"].iloc[i] / 1e9:.1f}B', fontsize = 9, \
ha = "right", va = "bottom", color = "black")

plt.grid(True, which = "both", axis = "both", linestyle = "-", \
linewidth = 0.7, color = "gray")
plt.xticks(df["Year"].unique().astype(int))

plt.title("Total Revenue over Years")
plt.xlabel("Year")
plt.ylabel("Total Revenue in Billions ($)")
plt.yscale("log")
plt.legend()


plt.show()
```
